```python
# Simple routine to run a query on a database and print the results:
def doQuery() :
    cur = myConnection.cursor()
    cur.execute("SELECT id, item_id, price from casas")
    for idnum,item_id, price in cur.fetchall():
        print(idnum,item_id, price)
```

```python
def insert_db(item_id, title, price, currency, latitude, longitude, barrio, direccion, permalink, total_area):

    #print("inserting --> "+title)
    
    mySql_insert_cols = "INSERT INTO casas (item_id, title, price, currency, latitude, longitude, barrio, direccion, permalink, total_area)"
    mySql_insert_vals = "values ('"+item_id+"','"+title+"',"+str(price)+",'"+currency+"',"+str(latitude)+","+str(longitude)+",'"+barrio+"','"+direccion+"','"+permalink+"',"+str(total_area)+")"

    cursor = myConnection.cursor()
    #print(mySql_insert_cols+mySql_insert_vals)
    cursor.execute(mySql_insert_cols+mySql_insert_vals)
    myConnection.commit()
        
```

```python
def printData(casa):
        print("-------------------")
        print(casa["id"])
        print(casa["title"])
        print(casa["price"])
        print(casa["currency_id"])
        print(casa["location"]["latitude"])
        print(casa["location"]["longitude"])
        print(casa["location"]["neighborhood"]["name"])
        print(casa["location"]["address_line"])
        print(casa["permalink"])
```

```python
def processCasa(j):
    if 'results' in j:
        for casa in j["results"]:    
            
            item_id = casa["id"]
            title = casa["title"]
            title = title.replace("'", "-")
            price = casa["price"]
            currency = casa["currency_id"]
            latitude = casa["location"]["latitude"]
            longitude = casa["location"]["longitude"]
            barrio = casa["location"]["neighborhood"]["name"]
            barrio = barrio.replace("'", "-")
            direccion = casa["location"]["address_line"]
            direccion = direccion.replace("'", "-")
            permalink = casa["permalink"]                                        
            #printData(casa)
            total_area = 0
            covered_area = 0
            
            for attr in casa["attributes"]:
                if attr["id"]=="TOTAL_AREA":
                    total_area = float(attr["value_name"][0:-2])
            
            for attr in casa["attributes"]:
                if attr["id"]=="COVERED_AREA":
                    covered_area = float(attr["value_name"][0:-2])

            if covered_area>total_area:
                total_area=covered_area
            
            insert_db(item_id, title, price, currency, latitude, longitude, barrio, direccion, permalink, total_area)
        
```

```python
def callAPI(offset, tipo, barrio):
    url = "https://api.mercadolibre.com/sites/MLA/search?category=MLA1459&9991459-AMLA_1459_2=9991459-AMLA_1459_2-MMLA12623&state=TUxBUEdSQWU4ZDkz&city=TUxBQ1ZJQ2E3MTQz&limit=50"+"&offset="+str(offset)+"&9991459-AMLA_1459_1="+tipo+"&neighborhood="+barrio
    r = requests.get(url)
    j = json.loads(r.content)
    return j
```

```python
def calc_offset(tipo, barrio):
    offset = 0
    j = callAPI(offset, tipo, barrio) #llama una vez para obtener el total
    total_rows = j["paging"]["total"] #obtiene la cantidad de casas
    print("Total:"+str(total_rows))
    return total_rows
```

```python
def procesa_lote(tipo, barrio):
    total_rows = calc_offset(tipo, barrio)
    llamadas = math.ceil(total_rows/50) #cantidad de veces a hacer call a la api
    for llamada in range(0,llamadas+1):
        offset = llamada*50
        print("offset: "+str(offset))
        j = callAPI(offset, tipo, barrio)
        processCasa(j)
```
